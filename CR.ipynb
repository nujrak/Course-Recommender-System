```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the page to scrape
url = 'https://www.coursera.org/courses'

# Send a GET request to the page
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the course cards
courses = soup.find_all('li', class_='ais-InfiniteHits-item')

# Open a CSV file to write the data
with open('coursera_courses.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'URL', 'Category', 'Sub-Category', 'Skills', 'Rating', 'Organization'])  # Write header

    # Write course details to the CSV file
    for course in courses:
        # Extract the required details
        title_element = course.find('h2', class_='color-primary-text card-title headline-1-text')
        url_element = course.find('a', class_='rc-DesktopSearchCard anchor-wrapper')
        category_element = course.find('a', class_='partner-name')
        skills_element = course.find('div', class_='card-info')
        rating_element = course.find('span', class_='ratings-text')
        organization_element = course.find('span', class_='partner-name')
        
        title = title_element.text.strip() if title_element else 'N/A'
        url = 'https://www.coursera.org' + url_element['href'] if url_element else 'N/A'
        category = category_element.text.strip() if category_element else 'N/A'
        sub_category = 'N/A'  # Assuming sub-category is not directly available
        skills = skills_element.text.strip() if skills_element else 'N/A'
        rating = rating_element.text.strip() if rating_element else 'N/A'
        organization = organization_element.text.strip() if organization_element else 'N/A'

        # Write the row to the CSV file
        writer.writerow([title, url, category, sub_category, skills, rating, organization])

print("Data has been written to coursera_courses.csv")
```
